2. EL atractor de Lorenz
>*¿Puede el aleteo de una mariposa en Brasil provocar un tornado en Texas?* <br>
En 1963, Edward Lorenz (1917-2008), muy interesado por el problema de la convección en la atmósfera terrestre,
simplifica drásticamente las ecuaciones Navier-Stokes de mecánica de fluidos, famosas por su complejidad, creando
así el célebre Atractor de Lorenz.
El atractor de Lorenz es un sistema de ecuaciones diferenciales autónomo, un sistema dinámico, que en principio
su autor lo propuso para tratar de comprender los fenómenos meteorológicos. Se encontró con que a pesar de que
los valores generados nunca se repiten y de que las condiciones iniciales pueden hacer variar completamente los
valores generados (de ahí su nombre de atractor extraño o caótico), el atractor toma una forma única y parece
conservar cierto orden, sus infinitas trayectorias nunca se cortan (pues eso implicaría que el sistema entraría en
un ciclo periódico), y también puede comprobarse cómo este objeto presenta la autosimilitud característica de los
fractales.
Las ecuaciones que describen el sistema son:<br>
<br>
$$ \frac{dx}{dt}= a(y − x)  \qquad \qquad (1)$$<br>
$$ \frac{dy}{dt}= x(b − z)-y \qquad \qquad (2)$$<br>
$$ \frac{dz}{dt}= xy-cz \qquad \qquad (3)$$<br>
>Donde $a$ es llamado el número de Prandtl y $b$ se llama el número de Rayleigh.<br>
Las constantes $(a, b, c) > 0$, pero es usualmente $a = 10$, $c = 8=3$ y $b$ es variado.<br>
El sistema exhibe un comportamiento caótico para $b = 28$ pero muestra ´orbitas periódicas para otros valores de $b$;
por ejemplo, con $b = 99:96$ se convierte en un nudo teórico llamado $T(3, 2)$.<br>
La forma de mariposa del atractor de Lorenz puede haber inspirado el nombre del efecto mariposa en la teoría del

(a).  Metodo de Euler:<br><br>
Aplicando el método de Euler al sistema de ecuaciones diferenciales:

$$ \frac{dx}{dt}= a(y − x)  \qquad \qquad (1)$$
$$ \frac{dy}{dt}= x(b − z)-y \qquad \qquad (2)$$
$$ \frac{dz}{dt}= xy-cz \qquad \qquad (3)$$<br>

$$ X=\begin{bmatrix} x\\  y \\ z  \end{bmatrix},\qquad X_o=\begin{bmatrix} x_o\\  y_o \\ z_o  \end{bmatrix},\qquad A=\begin{bmatrix} a\\ b\\c \end{bmatrix}  $$
$$   $$

$$\begin{bmatrix} \frac{dx}{dt} \\ \frac{dy}{dt}  \\ \frac{dz}{dt}  \end{bmatrix} = \begin{bmatrix} a(y − x) \\  x(b − z)-y  \\ xy-cz  \end{bmatrix} =F\begin{pmatrix}x\\  y \\ z \\ t \end{pmatrix}=F(X,t) $$



$$m= \frac{t_{max}-t_0}{n}$$


$$  t_k =t_{k-1}+m \quad donde\quad k=(1,2..,k,..,n)$$

Según este método una aproximación de $X$ es decir del vector $\begin{bmatrix} x\\  y \\ z  \end{bmatrix}$ en el tiempo $t_{k-1} es:$ 

$$  X_k =X_{k-1}+mF(X_{k-1},t_{k-1}) \quad es\quad decir: $$

$$  \begin{bmatrix} x\\  y \\ z  \end{bmatrix}_k =\begin{bmatrix} x\\  y \\ z  \end{bmatrix}_{k-1}+mF\begin{pmatrix}x\\  y \\ z \\ t \end{pmatrix}_{k-1} =\begin{bmatrix} x_{k-1}\\  y_{k-1} \\ z_{k-1}  \end{bmatrix}+m\begin{bmatrix} a(y_{k-1} − x_{k-1}) \\  x_{k-1}(b − z_{k-1})-y_{k-1}  \\ x_{k-1}y_{k-1}-cz_{k-1}  \end{bmatrix} $$

Para nuestro caso tenemos que $X_o=\begin{bmatrix} 0\\  1 \\ 1.05  \end{bmatrix}$, $t_{max}=100$ y $n=10000$, también para 
el atractor de Lorenz b=28, tenemos $A=\begin{bmatrix} 10\\ 28\\ \frac{8}{3} \end{bmatrix}$ y para el atractor de Lorenz b=96.96, tenemos que $A=\begin{bmatrix} 10\\ 99.96\\ \frac{8}{3} \end{bmatrix}$, entonces desarrollando el método usando Python, tenemos lo siguiente:





```python
import matplotlib.pyplot as plt # Importamos librería de graficación
import numpy as np  # Importamos numpy para manejo de arrays 

def atractor_lorenz_euler(A,XO,n,tmax,t_0): # Definimos subprograma de solución númerica de SEDO usando método de Euler
    m=(tmax-t_0)/n # Definimos incremento de tiempo m
    X=[XO[0]] # Definimos la primera estrada del vector x (como lista), que nos guardará cada valor aproximado de x en cada tiempo t
    Y=[XO[1]] # Definimos la primera estrada del vector y (como lista), que nos guardará cada valor aproximado de y en cada tiempo t
    Z=[XO[2]] # Definimos la primera estrada del vector z (como lista), que nos guardará cada valor aproximado de z en cada tiempo t
    t=[t_0]   # Definimos la primera estrada del vector t (como lista),  t en cada iteración
    
    for k in range (n): 
        t.append(t[-1]+m) #Agregamos el tiempo k-ésimo en el vector de tiempo
        X.append(X[-1]+m*(A[0]*(Y[-1]-X[-1]))) #Agregamos el valor  k-ésimo de x en el vector x
        Y.append(Y[-1]+m*(X[-1]*(A[1]-Z[-1])-Y[-1]))#Agregamos el valor k-ésimo de y en el vector y
        Z.append(Z[-1]+m*(X[-1]*Y[-1]-A[2]*Z[-1]))#Agregamos el valor k-ésimo z en el vector z
        
    return(X,Y,Z,t,m)

X1,Y1,Z1,t1,m1=atractor_lorenz_euler([10,28,8/3],[0,1,1.05],10000,100,0) 
# Guardamos en los vectores X1,Y1,Z1,t1, los valores de x,y,z en cada t del atractor de Lorenz para b=28
X2,Y2,Z2,t2,m2=atractor_lorenz_euler([10,99.96,8/3],[0,1,1.05],10000,100,0)
# Guardamos en los vectores X2,Y2,Z2,t2, los valores de x,y,z en cada t del atractor de Lorenz para b=99.96

plt.figure(figsize=(15,10)) # Generamos un "figure" y le damos un tamaño

plt.subplot(2,3,1) # Dividimos el figure en una matriz 2x3, y seleccionamos el elemento 1 (posición 1,1)
plt.plot(X1, Y1,'g', linewidth=0.5, linestyle="-.") # Damos valores en x(X1) y en Y(Y1) al elemento 1 de la matriz del figure
# también le damos un ancho de 0.5 a linea, color 'g'(green) y un estilo de línea "-."
plt.xlabel("x") # titulo del eje x
plt.ylabel("y") # titulo del eje y, similarmente se hace con el resto de subplots

plt.subplot(2,3,2)
plt.plot(X1, Z1,color='purple',linewidth=0.5, linestyle="-.")
plt.xlabel("z")
plt.ylabel("x")
plt.title("Atractor de Lorenz (EULER),  b=28", fontsize=20) # Título en grafica central para poner título del respectivo atractor

plt.subplot(2,3,3)
plt.plot(Y1, Z1,'r', linewidth=0.5, linestyle="-.")
plt.xlabel("z")
plt.ylabel("y")

plt.subplot(2,3,4)
plt.plot(X2, Y2,'blue', linewidth=0.5, linestyle="-.")
plt.xlabel("y")
plt.ylabel("x")

plt.subplot(2,3,5)
plt.plot(X2, Z2,'y',linewidth=0.5, linestyle="-.")
plt.xlabel("z")
plt.ylabel("x")
plt.title("Atractor de Lorenz (EULER), b=99.96", fontsize=20)

plt.subplot(2,3,6)
plt.plot(Y2, Z2,linewidth=0.5, linestyle="-.")
plt.xlabel("z")
plt.ylabel("y")
```

(b).  Odeint:<br><br>

```python
import numpy as np
from scipy.integrate import odeint # Importamos odeint, modulo para resolución de SEDO de scipy
import matplotlib.pyplot as plt

'''Para usar  odeint es necesario definir una función de dos parámetos, el primero un vector (en este caso como lista) que 
contenga las variables dependientes, en este caso x[0]=x, x[1]=y, x[2]=z, el segundo parámetro es la variable independiente
, en este caso es t'''
 
def f1(X,t): # Definimos una función f1, para usar odeint cuando b=28
    return [10*(X[1]-X[0]),X[0]*(28-X[2])-X[1],X[0]*X[1]-(8/3)*X[2]] # la función debe devolver un vector (lista)
# donde la primera entrada es dx/dt, la segunda dy/dt y la tercera dz/dt

def f2(X,t): # Definimos una función f2, para usar odeint cuando b=99.696
    return [10*(X[1]-X[0]),X[0]*(99.96-X[2])-X[1],X[0]*X[1]-(8/3)*X[2]]

# creamos vector con valores iniciales
f0 = [0,1,1.05]

# Unando numpy generamos vector (un arreglo) de los incrementos de tiempo
t = np.linspace(0,100,10000)


# solucionamos ambos sistemas de ecuaciones (cuando b=28 y cuando b=99.96) y la guardamos en una matriz sol que contiene
# la aproximacion de los valores de x, y, z en cada tiempo t
sol1 = odeint(f1,f0,t)
sol2 = odeint(f2,f0,t)

# dicha matriz "sol" tiene en la primera columna todos los valores de x (sol[:0]) es decir si queremos tomar el valor
# de los x seleccionamos todos los elementos de la columna 0

plt.figure(figsize=(15,10))# Generamos un "figure" y le damos un tamaño

plt.subplot(2,3,1) # Dividimos el figure en una matriz 2x3, y seleccionamos el elemento 1 (posición 1,1)
plt.plot(sol1[:,0],sol1[:,1],'g', linewidth=0.5, linestyle="-.")# Damos valores en x y en y al elemento 1 de la matriz del figure
# también le damos un ancho de 0.5 a linea, un color 'g'(green) y un estilo de línea "-."
plt.xlabel("x")# titulo del eje x
plt.ylabel("y")# titulo del eje y, similarmente se hace con el resto de subplots

plt.subplot(2,3,2)
plt.plot(sol1[:,0], sol1[:,2],color='purple',linewidth=0.5, linestyle="-.")
plt.xlabel("z")
plt.ylabel("x")
plt.title("Atractor de Lorenz (ODEINT), b=28", fontsize=20)

plt.subplot(2,3,3)
plt.plot(sol1[:,1], sol1[:,2],'r', linewidth=0.5, linestyle="-.")
plt.xlabel("z")
plt.ylabel("y")

plt.subplot(2,3,4)
plt.plot(sol2[:,0], sol2[:,1],'blue', linewidth=0.5, linestyle="-.")
plt.xlabel("y")
plt.ylabel("x")

plt.subplot(2,3,5)
plt.plot(sol2[:,0], sol2[:,2],'y',linewidth=0.5, linestyle="-.")
plt.xlabel("z")
plt.ylabel("x")
plt.title("Atractor de Lorenz (ODEINT), b=99.96", fontsize=20)

plt.subplot(2,3,6)
plt.plot(sol2[:,1], sol2[:,2],linewidth=0.5, linestyle="-.")
plt.xlabel("z")
plt.ylabel("y")
```
